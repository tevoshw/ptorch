# Tensor like

##### `Tensor.x_like` it's a way to create/convert/copy tensors in any dimensions, `tensor.x_like(tensor)` gonna create the tensor with the especify X with the size, this function copy e exist tensor1 and create a new with the same dimensions and shape like the tensor1, just change the numbers,  like the exemple below:


```python
import torch

# Create tensors with 0 and 1
scalar = torch.rand(1)
vector = torch.rand(2,)
matrix = torch.rand(3,3)
tensor = torch.rand(4,4,4)
print("Orignals tensors")
print(scalar)
print(vector)
print(matrix)
print(tensor)
print("\n\n\n")

# Convert/Create to zeros

print("Converrt the TENSORS RAND TO TENSORS ZEROS")
scalar = torch.zeros_like(scalar) 
vector = torch.zeros_like(vector) 
matrix = torch.zeros_like(matrix) 
tensor = torch.zeros_like(tensor) 
print(scalar)
print(vector)
print(matrix)
print(tensor)
print("\n\n\n")



# Convert/Create to ones
print("Converrt the TENSORS ZEROS TO TENSORS ONES")
scalar = torch.ones_like(scalar) 
vector = torch.ones_like(vector) 
matrix = torch.ones_like(matrix) 
tensor = torch.ones_like(tensor) 
print(scalar)
print(vector)
print(matrix)
print(tensor)
print("\n\n\n")

# Convert/Create to randint
print("Converrt the TENSORS ONES TO TENSORS RANDINT")
scalar = torch.randint_like(scalar, 1, 10) 
vector = torch.randint_like(vector, 1, 10) 
matrix = torch.randint_like(matrix, 1, 10) 
tensor = torch.randint_like(tensor, 1, 10) 
print(scalar)
print(vector)
print(matrix)
print(tensor)
```
