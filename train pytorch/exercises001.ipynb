## Create tensors

```python
import torch 

tensor = torch.randn((5, 5), dtype = torch.float16)
tensor_ajustado = torch.where(tensor < 0, 0, tensor)

print(f"Tensor antes de se ajustar:\n{tensor}")
print(f"\n\nTensor após ser ajustado:\n{tensor_ajustado}")
```

```python
tensor = torch.arange(1, 26, dtype= torch.float16).reshape(5, 5)
print(f"Tensor aspiral:\n{tensor}")
```

```python
tensor = torch.randint(0, 10, (4, 4), dtype = torch.float16)
print(f"Tensor antes dos ajustes com números de 0 à 10:\n{tensor}")

media = torch.mean(tensor)
tensor = torch.where(tensor < media, 0, 1)
print(f"\n\nTensor depois dos ajustes:\n{tensor}")
```

```python
matrix = torch.eye(6).type(torch.float16)
print(f"Criando a matrix identidade:\n{matrix}")

matrix[torch.arange(6), torch.arange(6)] = torch.randint(1, 11, (6,), dtype=torch.float16)
print(f"\n\nMatrix ajustada para a identidade ser números aleatórios:\n{matrix}")
```
